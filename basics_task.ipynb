```python
#1
def find_average(numbers):
  sum = 0
  for i in numbers:
    sum += i

  return sum / len(numbers)

find_average([20, 30, 40])
```

```python
#3
def print_pattern(range_number):
  lst = []
  for i in range(range_number):
    lst.append(i)
    print(*lst)


print_pattern(10)
```

```python
#4
def print_multi(number):
  for i in range(10):
    print(f"{number} x {i} = {number * i}")

print_multi(4)
```

```python
#5
def count_length(number):
  return len(str(number))

print(count_length(23423452))
```

```python
#6
def is_anagram(str1, str2):
  return "anagram" if sorted(str1) == sorted(str2) else "no anagram"

print(is_anagram("sdfsd", "sdfsd"))
```

```python
#7
def calculation(n1, n2):
  return f"Addition: {n1 + n2} \nSubtraction: {n1 - n2}"

print(calculation(12, 15))
```

```python
#8
def show_employee(name, salary = 9000):
  return name, salary

print(show_employee("Jasur"))
```

```python
#9
def find_even():
  lst = []
  for i in range(4, 30):
    if i not in lst and i % 2 == 0:
      lst.append(i)

  return lst

print(find_even())
```
