```python
#Multiply a 5x3 matrix by a 3x2 matrix (real matrix product)
import numpy as np
arr1 = np.arange(15).reshape(5,3)
arr2 = np.arange(6).reshape(3,2)
result = np.dot(arr1, arr2)
print(result)
```

```python
#  How to find common values between two arrays?
import numpy as np
arr1 = np.arange(15).reshape(5,3)
arr2 = np.arange(6).reshape(3,2)
commons = np.intersect1d(arr1, arr2)
print(commons)
```

```python
# Create a 5x5 matrix with row values ranging from 0 to 4
import numpy as np
arr = np.zeros((5,5), dtype=int)
arr = arr + np.arange(5)
print(arr)
```

```python
# Create a vector of size 10 with values ranging from 0 to 1, both excluded
import numpy as np
arr = np.linspace(0,1, 12)
print(arr[1:-1])
```

```python
# Consider two random arrays A and B, check if they are equal
import numpy as np
A = np.random.randint(0,5, 10, dtype=int)
B = np.random.randint(0,5, 10, dtype=int)
is_equal = np.array_equal(A, B)
print(is_equal)
```
