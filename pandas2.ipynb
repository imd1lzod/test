```python
import pandas as pd

data = {
    "user_id": [1,1,2,2,3,3,3],
    "category": ["A","B","A","A","B","B","A"],
    "amount": [100,200,150,300,400,100,50],
    "date": pd.to_datetime([
        "2024-01-01","2024-01-05",
        "2024-01-02","2024-01-07",
        "2024-01-03","2024-01-08","2024-01-10"
    ])
}
df = pd.DataFrame(data)

df = df.sort_values("date").set_index("date")

end = df.index.max()
start = end - pd.Timedelta(days=6)

filtered_data = df.loc[start:end]

total_last_7_days = filtered_data["amount"].sum()

print(filtered_data)
print("Last 7 days total:", total_last_7_days)
```
